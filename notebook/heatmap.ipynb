```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data with 10 variables and 100 observations
np.random.seed(42)  # For reproducibility
data = np.random.rand(100, 10)

# Compute the correlation matrix
correlation_matrix = np.corrcoef(data, rowvar=False)

# Create a heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", cbar=True, square=True,
            xticklabels=[f'Var{i+1}' for i in range(10)],
            yticklabels=[f'Var{i+1}' for i in range(10)])
plt.title("10x10 Correlation Matrix Heatmap")
plt.show()
```
